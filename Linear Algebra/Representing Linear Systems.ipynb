## 1 - System of Linear Equations

A **System of linear equations** is a collection of one or more linear equations involving same variables. For example:

$$
\begin{align*}
x+y=4 \\ 4x+2y=5 \end{align*}
$$

It is a system of two linear equations with two unknown variables $x$ and $y$.

If the Linear System has unique solution, it is **Non-Singular**, or else if the System has No solution or Infinite Solutions, it is **Singular**

<a name = '2'></a>
## 2 - System of Linear Equations using Matrices

Above System of Linear Equations are represented using Matrices as:

Coefficient matrix:

$$
\begin{bmatrix}
1 & 1  \\
4 & 2
\end{bmatrix}
$$

Output Vector:

$$
\begin{bmatrix}
4 \\
5
\end{bmatrix}
$$

Each row coefficinet matrix represents an equation in the system. First column represents coefficients of $x_1$ in the system, the second column represents the coefficients of $x_2$.
Output Vector represents the constant values on the right side of the equation.

```python
A = np.array([
    [1,1],
    [4,2]
])

b = np.array([4,5])

print("Coefficient Matrix: ")
print(A)
print("\nOutput Vector: ")
print(b)
```

```python
print(f"Shape of Coefficient Matrix: {np.shape(A)}")
print(f"Shape of Output Vector: {np.shape(b)}")
```

```python
solution = np.linalg.solve(A,b)
print(f"Solution fo the equations: {solution}")
```

## 3 - Evaluating Determinant of a Matrix to determine its Singularity

If the determinant of a matrix is 0, then the System of Equations is **Singular** (Linear Dependent).
else, System of Equations is **Non-Singular** (Not Linear Dependent)

```python
d = np.linalg.det(A)
print(f"Determinant of matrix A: {d}")
```

```python
X = np.array([
    [1,1],
    [2,2]
])

d = np.linalg.det(X)
print(f"Determinant of matrix X: {d}")
```

A is Non Singular and X is Singular

## 4 - Visualizing System of Linear Equations


```python
# Function to plot System of Linear Equations

def plot_lines(A):

  plt.figure(figsize=(5,5))
  x = np.linspace(-10,10,100)

  y1 = (A[0,2]-A[0,0]*x)/A[0,1]
  y2 = (A[1,2]-A[1,0]*x)/A[1,1]

  plt.plot(x,y1,label='Equation 1')
  plt.plot(x,y2,label='Equation 2')
  plt.xlabel('x')
  plt.ylabel('y')

  X = A[:,0:2]
  y = A[:,-1]
  if np.linalg.det(X)!=0:
    sol = np.linalg.solve(X,y)
    plt.plot(sol[0],sol[1],'ro',label='Solution')

  plt.grid(True)
  plt.title('Linear Equations in 2D')
  plt.legend()
  plt.show()
```

<a name='4.1'></a>
### 4.1 - System of Linear Equations with Unique Solution

Here we are considering the equations:

$$
\begin{align*}
x + y = 4 \\
4x + 2y = 5
\end{align*}
$$

```python
A = np.array([
    [1,1,4],
    [4,2,5]
])

plot_lines(A)

```

### 4.2 - System of Linear Equations with No solution

Here we are considering equations:

$$
\begin{align*}

2x+2y=20
\end{align*}
$$

```python
B = np.array([
    [1,1,4],
    [2,2,20]
])

plot_lines(B)
```

### 4.3 - System of Linear Equations with Infinite Solutions

Here we are considering equations:

$$
